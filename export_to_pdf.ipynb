# Convert to PDF

Когда все этапы анализа данных завершены, отчет можно экспортировать:<br>
* Если вам нужны заголовки или текст в отчете, то пишите его в ячейках ноутбука, сменив формат Code на Markdown. 
* Для экспорта можно использовать терминал, запуская там вторую строку без восклицательного знака, либо можно запустить код, написанный ниже, в ячейке ноутбука. 
* Чтобы исключить ячейки с кодом (не экспортировались в pdf) используем параметр TemplateExporter.exclude_input=True.<br> 
* Также, при запуске этой ячейки код выдает стандартный поток (standard output) и, чтобы в отчете его не было видно, в начале ячейки нужно написать %%capture.
